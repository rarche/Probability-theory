***1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.***

```python
M = (200+800)/2
D = ((800-200)**2)/12
print(f'Среднее значение = {M}; дисперсия = {D}')
```

```python
import numpy as np
D = 0.2
a = 0.5
b_sub_a = np.sqrt(D*12)
b = b_sub_a + a
M = (a + b)/2
print(f'Правая граница равномерно распределенной величины В = {round(b,2)}, среднее значение равномерно распределенной величины В = {round(M,2)} ')
```

***4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.***

```python
def standart_normal_distr(x):
    return (x-174)/8

s_182 = standart_normal_distr(182) #1.0 - 0.8413
p_182 = 0.8413
s_190 = standart_normal_distr(190) #2.0 - 0.9772
p_190 = 0.9772
s_166 = standart_normal_distr(166) #-1.0 - 0.1587
p_166 = 0.1587
s_158 = standart_normal_distr(158) #-2.0 - 0.0228
p_158 = 0.0228
s_150 = standart_normal_distr(150) #-3.0 - 0.0014
p_150 = 0.0014
s_198 = standart_normal_distr(198) #3.0 - 0.9986
p_198 = 0.9986
print(f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост: \nа). больше 182 см: {round(1 - p_182,3)} \nб). больше 190 см : {round(1 - p_190,3)} \nв). от 166 см до 190 см: {round(p_190,3) - round(p_166,3)} \nг). от 166 см до 182 см: {round((p_182 -p_166),3)} \nд). от 158 см до 190 см: {round(p_190,3) - round(p_158,3)} \nе). не выше 150 см или не ниже 190 см: {round(p_150,3) + round((1 - p_190),3)} \nё). не выше 150 см или не ниже 198 см: {round(p_150,3) + round((1 - p_198),3)} \nж). ниже 166 см.: {round(p_166, 3)}')
```
